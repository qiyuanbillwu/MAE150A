```julia
using Plots
```

```julia
β = π/4
R = 2
Γ = 1.0
```

```julia
F(z) = im*Γ/(2*pi)*log(z^(π/β)+im*R^(π/β))-im*Γ/(2*pi)*log(z^(π/β)-im*R^(π/β))-im*Γ/(2*pi)*log(z^(-π/β)-im*R^(π/β))+im*Γ/(2*pi)*log(z^(-π/β)+im*R^(π/β))
```

```julia
xg = range(-3,3,length=300)
yg = range(-3,3,length=300)
zg = [x + im*y for x in xg, y in yg];
```

```julia
plot(xg,yg,imag(F.(zg)),ratio=1,color=:black)
```
