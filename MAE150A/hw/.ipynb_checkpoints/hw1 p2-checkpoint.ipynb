```julia
using FluidDynTools
```

```julia
using Plots
```

```julia
using SpecialFunctions
```

```julia
u(y,nu,t) = 1-erf(y/(2*sqrt(nu*t)))
```

```julia
nu = KinematicViscosity(1)
```

```julia
t1 = 0.1u"s"
t2 = 1u"s"
t3 = 10u"s"
```

```julia
y1 = range(0u"m",2u"m",length=101)
y2 = range(0u"m",10u"m",length=101)
y3 = range(0u"m",20u"m",length=101)
```

```julia
v1 = u.(y1,nu,t1)
v2 = u.(y2,nu,t2)
v3 = u.(y3,nu,t3)
```

```julia
plot(v1,y1,xlim=(0,Inf),ylim=(0,Inf),xlabel=L"u(y)",ylabel=L"y")
```

```julia
plot(v2,y2,xlim=(0,Inf),ylim=(0,Inf),xlabel=L"u(y)",ylabel=L"y")
```

```julia
plot(v3,y3,xlim=(0,Inf),ylim=(0,Inf),xlabel=L"u(y)",ylabel=L"y")
```
